### Exercise 1:
Create a simple line plot for `y = x` for `x` values ranging from 0 to 10.

```python
# Exercise 1
```

````python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
````

````python
x = np.linspace(0,10)
y = x


plt.figure(figsize=(10, 14))


plt.plot(x, y, 'r')


plt.xlabel('x')
plt.ylabel('y')


plt.title(r"Plot of $y=x$")
plt.show()
````

### Exercise 3:
Create a scatter plot with 50 random points where x and y values are between 0 and 10.

```python
# Exercise 3
```

````python
plt.figure(figsize=(10, 6), dpi=80)

n = 50
X = np.random.uniform(0, 10, n)
Y = np.random.uniform(0, 10, n)

# Create a scatter plot
plt.scatter(X, Y, s=75, alpha=0.5)

# Set axis limits
plt.xlim(0, 10)
plt.ylim(0, 10)

# Set x and y ticks
plt.xticks(np.arange(0, 11, 1))
plt.yticks(np.arange(0, 11, 1))

# Show the plot
plt.title('Scatter Plot of 50 Random Points')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.show()
````

### Exercise 9:
Create a subplot with 2 rows and 1 column. Plot `y1 = x^2` in the first subplot and `y2 = x^3` in the second.

```python
# Exercise 9
```

---

````python
x = np.linspace(0, 10, 100)
y = x ** 2

fig, axes = plt.subplots(1, 2, figsize=(8,4), dpi=100)

# plot subplot 1
axes[0].plot(x, x**2, color="green", label="y = x**2")
axes[0].legend(loc=2); # upper left corner
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('Plot of y=x^2')

# plot subplot 2
axes[1].plot(x, x**3, color="blue", label="y = x**3")
axes[1].legend(loc=2); # upper left corner
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].set_title('Plot of y=x^3')

# `fig.tight_layout()` automatically adjusts the positions of the axes on the figure canvas so that there is no overlapping content
# comment this out to see the difference
fig.tight_layout()

plt.show()
````

````python
x = np.linspace(0, 10, 100)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 8), dpi=100)


ax1.plot(x, x**2, 'r', label=r"$y = x^2$")
ax1.set_xlabel(r'$x$', fontsize=18)
ax1.set_ylabel(r'$y$', fontsize=18)
ax1.set_title(r'Plot of $y = x^2$', fontsize=18)
ax1.legend(loc='upper left')



ax2.plot(x, x**3, 'b', label=r"$y = x^3$")
ax2.set_xlabel(r'$x$', fontsize=18)
ax2.set_ylabel(r'$y$', fontsize=18)
ax2.set_title(r'Plot of $y = x^3$', fontsize=18)
ax2.legend(loc='upper left')



plt.tight_layout()


plt.show()
````

````python
x = np.linspace(0, 10, 100)
y1 = x**2
y2 = x**3

plt.figure(figsize=(8, 8))

plt.subplot(2, 1, 1) 
plt.plot(x, y1, 'r', label=r"$y = x^2$")
plt.title(r"Plot of $y = x^2$")
plt.xlabel('x')
plt.ylabel('y1')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(x, y2, 'b', label=r"$y = x^3$")
plt.title(r"Plot of $y = x^3$")
plt.xlabel('x')
plt.ylabel('y2')

plt.legend()

plt.tight_layout()

plt.show()
````

```python
# Exercise 14
```

---

````python
x = np.linspace(0, 10, 100)
y = x**2

plt.figure(figsize=(8, 6), dpi=100)

# Plot the filled area under the curve
plt.fill_between(x, y, color='skyblue', alpha=0.4, label=r'Area under $y = x^2$')

# Plot the curve
plt.plot(x, y, 'r', label=r"$y = x^2$")

# Set labels and title
plt.xlabel(r'$x$', fontsize=18)
plt.ylabel(r'$y$', fontsize=18)

plt.show()
````

### Exercise 17:
Create a plot with multiple subplots in a 2x2 grid. Plot `y1 = x`, `y2 = x^2`, `y3 = x^3`, and `y4 = x^4`.

```python
# Exercise 17
```

---

````python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values
x = np.linspace(0, 10, 100)

# Create a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 10), dpi=100)

# Plot y1 = x
axes[0, 0].plot(x, x, label=r'$y = x$', color='r')
axes[0, 0].set_title(r'Plot of $y = x$', fontsize=14)
axes[0, 0].set_xlabel(r'$x$', fontsize=12)
axes[0, 0].set_ylabel(r'$y$', fontsize=12)
axes[0, 0].grid(True)

# Plot y2 = x^2
axes[0, 1].plot(x, x**2, label=r'$y = x^2$', color='g')
axes[0, 1].set_title(r'Plot of $y = x^2$', fontsize=14)
axes[0, 1].set_xlabel(r'$x$', fontsize=12)
axes[0, 1].set_ylabel(r'$y$', fontsize=12)
axes[0, 1].grid(True)

# Plot y3 = x^3
axes[1, 0].plot(x, x**3, label=r'$y = x^3$', color='b')
axes[1, 0].set_title(r'Plot of $y = x^3$', fontsize=14)
axes[1, 0].set_xlabel(r'$x$', fontsize=12)
axes[1, 0].set_ylabel(r'$y$', fontsize=12)
axes[1, 0].grid(True)

# Plot y4 = x^4
axes[1, 1].plot(x, x**4, label=r'$y = x^4$', color='m')
axes[1, 1].set_title(r'Plot of $y = x^4$', fontsize=14)
axes[1, 1].set_xlabel(r'$x$', fontsize=12)
axes[1, 1].set_ylabel(r'$y$', fontsize=12)
axes[1, 1].grid(True)

plt.tight_layout()

# Show the plot
plt.show()
````

### Exercise 19:
Add error bars to a bar plot with `categories = ['A', 'B', 'C', 'D']` and `values = [3, 7, 5, 9]`, assuming the error is 1 for each value.

```python
# Exercise 19
```

---

````python
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 5, 9]

errors = np.ones(len(values)) 


plt.figure(figsize=(8, 6))
plt.bar(categories, values, yerr=errors, color='blue', capsize=5, alpha=0.7)


plt.xlabel('Categories', fontsize=14)
plt.ylabel('Values', fontsize=14)
plt.title('Bar Plot with Error Bars', fontsize=16)
plt.grid(axis='y')


plt.show()
````

### Exercise 20:
Plot a sine wave using a dashed line and a cosine wave using a solid line on the same plot.

```python
# Exercise 20
```

---

````python

x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)


y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(10, 6))


plt.plot(x, y_sin, label='Sine Wave', linestyle='--', color='blue')


plt.plot(x, y_cos, label='Cosine Wave', linestyle='-', color='red')


plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.title('Sine and Cosine Waves', fontsize=16)
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')  # Horizontal line at y=0
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')  # Vertical line at x=0


plt.legend()

plt.show()
````
